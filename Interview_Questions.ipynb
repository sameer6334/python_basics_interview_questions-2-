## 5. How does Python handle mutability and immutability? 3

### [Data Science, Data Analytics, Machine Learning Engineer], [Google, Zomato], 

```python
'''
In Python, the concept of mutability and immutability is important for understanding how objects behave when modified. Mutable objects can be changed after creation, whereas immutable objects cannot. This distinction has significant implications for memory management, performance, and safety when handling data, especially in large-scale or multi-threaded applications.

Behavior with Variables:-

MUTABLE OBJECTS:-

list1 = [1, 2, 3]
list2 = list1  # list2 points to the same object as list1
list2[0] = 10
print(list1)  # Output: [10, 2, 3] (list1 is modified too)

IMMUTABLE OBJECTS:-

str1 = "hello"
str2 = str1  # str2 points to the same object as str1
str2 = "world"  # This creates a new string object for str2
print(str1)  # Output: "hello" (str1 remains unchanged)
print(str2)  # Output: "world"
'''

```

## 10. What is the significance of conditionals in Python? 2

### [Business Analyst, Data Science], [Flipkart, Oracle]

```python
'''
1. Decision Making:-

Conditionals enable decision-making in a program, allowing it to behave differently under different circumstances. For example, a program might perform one action if a user is logged in and another if the user is not.

2. Control Flow:-

Conditionals are fundamental for defining the control flow of a program. They help in implementing various logic branches, loops, and decision paths. Without conditionals, all code would execute in a linear fashion, without the flexibility to make decisions at runtime.

3. Optimizing Code:-
Conditionals allow you to optimize your program’s behavior by executing code only when necessary. For example, certain computations or database queries can be skipped if the condition doesn’t require them.

4. Error Handling:-

Conditionals are often used to handle edge cases or potential errors in a program, ensuring that the program behaves correctly in all scenarios.

5. Flow of Execution in Loops:-

Conditionals are often used in loops to control how many times a loop executes, or when to exit the loop.

6. Enabling Complex Logic:-

You can combine multiple conditions using logical operators (and, or, not) to implement more complex decision-making in your code.

7. Customizing Program Behavior:-

Conditionals help customize the program’s output or behavior based on user input, external data, or other dynamic factors. This makes programs more flexible and interactive.
'''
```

## 13. How do you use break, continue, and pass in Python loops? 2

### [Data Analytics, Machine Learning Engineer], [Byju’s, Capgemini]

```python
'''
-> 1. break Statement
   Purpose: The break statement is used to exit or terminate a loop prematurely, regardless of whether the loop has finished iterating through all its     elements or not.

   When to use: You can use break when you want to stop the loop based on a specific condition, such as finding a particular item in a list or             encountering an error.


-> 2. continue Statement
   Purpose: The continue statement is used to skip the current iteration of the loop and move to the next iteration. It doesn't terminate the loop, but    it causes the rest of the code inside the loop to be skipped for the current iteration.

   When to use: You can use continue when you want to skip over certain parts of the loop under specific conditions, but continue the loop for the         remaining items.

-> 3. pass Statement
   Purpose: The pass statement is a no-op (no operation). It doesn't do anything and is typically used as a placeholder where a statement is               syntactically required, but you don't want to execute any code. It can be used in loops, functions, classes, and conditionals.

   When to use: You can use pass in situations where you need to write a loop, function, or condition, but you don't want to implement any action yet      (or you want to skip the current iteration without affecting the loop flow).
'''
```
